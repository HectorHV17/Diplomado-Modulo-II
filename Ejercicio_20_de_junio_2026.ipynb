## Ejercicio 1.
Se lanza una moneda equilibrada 10 veces.

Sea $X:$ el número de águilas obtenidas.
Calcula:  $P(X=k)$ y grafica la distribución de probabilidad. Además, calcula: $P(X=5)$ y $P(X \geq 7)$


**Sol:**

Sabemos que el experimento es lanzar una moneda equilibrada 10 veces.  
Como la moneda es equilibrada, cada lanzamiento es independiente, y la probabilidad de tener un exito es de 0.5, por lo tanto:

$$
X \sim Bin(10,0.5)
$$

Así, la función de densidad es:

$$
P(X=k)=\binom{10}{k}(0.5)^k(0.5)^{10-k}, \quad k=0,1,\ldots,10
$$

````python
#Importamos las librerias
import numpy as np
import numpy.random as npr # para generar numeros aleeatorios
import matplotlib.pyplot as plt #Para graficar
import scipy.stats as sps # distribuciones
from scipy.stats import binom
````

Calculamos las probabilidades:

````python
# Probabilidad de que haya 5 aguilas
# x=5, n=10, p=0.5
P5 = binom.pmf(5, 10, 0.5)
print("la probabilidad de que haya 5 aguilas es de", P5)

# Probabilidad de que haya 7 o mas aguilas
# P(X >= 7) = 1 - P(X <= 6)
P7 = 1 - binom.cdf(6, 10, 0.5)
print("la probabilidad de que haya 7 o mas aguilas es de", P7)
````

Realizamos la grafica:

````python
#Parametros
n = 10
p = 0.5
#Valores posibles de X
X = np.arange(0, n+1)
#Probabilidad
prob = binom.pmf(X, n, p)
#Grafica
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.bar(X, prob, color='#4A90E2', edgecolor='black')
plt.title('Distribucion Binomial (n=10, p=0.5)')
plt.xlabel('Número de águilas')
plt.ylabel('Probabilidad')
plt.grid(True)
#Volvemos a colocar las probabilidades, pero ahora que se vean en la grafica

P_X_5_1 = binom.pmf(5, n, p)
P_X_geq_7_1 = 1 - binom.cdf(6, n, p)

print(f'Ejercicio 1: P(X=5)={P_X_5_1:.4f}, P(X>=7)={P_X_geq_7_1:.4f}')
````

**Interpretación:**

Sabemos que la variable aleatoria $X$ representa el número de águilas obtenidas al lanzar una moneda equilibrada 10 veces.

La probabilidad de obtener exactamente 5 águilas es:

$$
P(X=5)=0.2461
$$

es decir, un 24.61\%.

La probabilidad de obtener al menos 7 águilas es:

$$
P(X\geq 7)=0.1719
$$

por lo que existe alrededor de un 17.19\% de probabilidad de obtener 7, 8, 9 o 10 águilas en los 10 lanzamientos.

Así, es más probable de obtener es alrededor de 5 águilas.

## Ejercicio 2.

Un examen tiene 12 preguntas. Cada pregunta tiene 4 opciones y solo una es correcta. Un estudiante responde todas las preguntas al azar.

Sea $X$ el número de respuestas correctas. Calcula: $ P(X=k)$ y grafica la distribución de probabilidad. Además, calcula: $P(X=3)$ y $P(X \geq 6)$


**Sol:**

Sabemos que el examen tiene 12 preguntas y cada pregunta tiene 4 opciones, por lo tanto la probabilidad de éxito en cada pregunta es \(0.25\).  

Como cada pregunta es independiente, tenemos que:

$$
X \sim Bin(12,0.25)
$$

Así, la función de densidad es:

$$
P(X=k)=\binom{12}{k}(0.25)^k(0.75)^{12-k}, \quad k=0,1,\ldots,12
$$

Calculamos las probabilidades:


````python
#Importamos las librerias
import numpy as np
import numpy.random as npr # para generar numeros aleeatorios
import matplotlib.pyplot as plt #Para graficar
import scipy.stats as sps # distribuciones
from scipy.stats import binom
````

````python
# Probabilidad de que que tenga 3 respuestas correctas
# x=3, n=12, p=0.25
P3 = binom.pmf(3, 12, 0.25)
print("la probabilidad de que tenga 3 respuestas correctas es de", P3)

# Probabilidad de que tenga 6 o mas respuestas correctas
# P(X >= 6) = 1 - P(X <= 5)
P6 = 1 - binom.cdf(5, 12, 0.25)
print("la probabilidad de que tenga 6  o mas respuestas correctas es de", P6)
````

Realizamos la grafica

````python
#Parametros
n2 = 12
p2 = 0.25
#Valores posibles de X
X2 = np.arange(0, n2+1)
#Probabilidad
prob2 = binom.pmf(X2, n2, p2)
#Grafica
plt.figure(figsize=(10,6))
plt.subplot(2, 2, 2)
plt.bar(X2, prob2, color='#2ECC71', edgecolor='black')
plt.title('Distribución Binomial (n=12, p=0.25)')
plt.xlabel('Respuestas correctas')
plt.ylabel('Probabilidad')
plt.grid(True)
#Volvemos a colocar las probabilidades, pero ahora que se vean en la grafica
P_X_3_2 = binom.pmf(3, n2, p2)
P_X_geq_6_2 = 1 - binom.cdf(5, n2, p2)
print(f'Ejercicio 2: P(X=3)={P_X_3_2:.4f}, P(X>=6)={P_X_geq_6_2:.4f}')
````

**Interpretación:**

Sabemos que la variable aleatoria $X$ representa el número de respuestas correctas al contestar 12 preguntas al azar.

Así, la probabilidad de acertar exactamente 3 preguntas es aproximadamente:

$$
P(X=3)=0.2581
$$

Esto significa que hay alrededor de un 25.81\% de probabilidad de obtener 3 aciertos.

Por otro lado, la probabilidad de acertar 6 o más preguntas es:

$$
P(X\geq 6)=0.0544
$$

lo que indica que es muy poco probable obtener 6 o más aciertos contestando completamente al azar.

## Ejercicio 3.

En promedio llegan 3 mensajes por minuto a un chat. Sea $X$ el número de mensajes que llegan en un minuto. Calcular: $P(X=k)$ y graficar la distribución de probabilidad. Además, calcular: $P(X=2)$ y $P(X \geq 5)$



**Sol:**

Sabemos que en promedio llegan 3 mensajes en un intervalo de un minuto a un chat, por lo tanto:

$$
X \sim \text{Poisson}(3)
$$

Así, la función de densidad es:

$$
P(X=k)=\frac{e^{-3}(3)^k}{k!}, \quad k=0,1,2,\ldots
$$

Calculamos las probabilidades:

````python
#Importamos las librerias
import numpy as np
import numpy.random as npr # para generar numeros aleeatorios
import matplotlib.pyplot as plt #Para graficar
import scipy.stats as sps # distribuciones
from scipy.stats import poisson
````

````python
# Probabilidad de que lleguen 2 mensajes en un minuto
# x=2, lambda=3
P2 = poisson.pmf(2, 3)
print("la probabilidad de que lleguen 2 mensajes en un minuto es de", P2)

# Probabilidad de que lleguen 5 o mas mensajes en un minuto
# P(X >= 5) = 1 - P(X <= 4)
P_5 = 1 - poisson.cdf(4, 3)
print("la probabilidad de que lleguen 5 o mas mensajes en un minuto es de", P_5)
````

Realizamos la grafica:


````python
#Parametros
lambda3 = 3
#Valores posibles de X
X3 = np.arange(0, 10)
#Probabilidad
prob3 = poisson.pmf(X3, lambda3)
#Grafica
plt.figure(figsize=(10,6))
plt.subplot(2, 2, 3)
plt.bar(X3, prob3, color='#E74C3C', edgecolor='black')
plt.title('Distribución de Poisson (λ=3)')
plt.xlabel('Mensajes por minuto')
plt.ylabel('Probabilidad')
plt.grid(True)
#Volvemos a colocar las probabilidades, pero ahora que se vean en la grafica
P_X_2_3 = poisson.pmf(2, lambda3)
P_X_geq_5_3 = 1 - poisson.cdf(4, lambda3)
print(f'Ejercicio 3: P(X=2)={P_X_2_3:.4f}, P(X>=5)={P_X_geq_5_3:.4f}')
````

**Interpretación:**

Sabemos que la variable aleatoria $X$ representa el número de mensajes que llegan en un minuto.

La probabilidad de recibir exactamente 2 mensajes es:

$$
P(X=2)=0.2240
$$

Esto significa que hay aproximadamente un 22.40\% de probabilidad de que lleguen 2 mensajes en un minuto.

Por otro lado, la probabilidad de recibir 5 o más mensajes es:

$$
P(X\geq 5)=0.1847
$$

lo que indica que es menos común que lleguen muchos mensajes en un minuto.

## Ejercicio 4.

Un libro tiene en promedio 1.5 errores por página. Sea $X$ el número de errores en una página. Calcular: $P(X=k)$ y grafica la distribución de probabilidad. Además, calcula: $P(X=0)$ y $P(X \leq 2)$

**Sol:**

Sabemos que un libro tiene en promedio 1.5 errores por página, por lo tanto:

$$
X \sim \text{Poisson}(1.5)
$$

Así, la función de probabilidad es:

$$
P(X=k)=\frac{e^{-1.5}(1.5)^k}{k!}, \quad k=0,1,2,\ldots
$$

Calculamos las probabilidades:


````python
#Importamos las librerias
import numpy as np
import numpy.random as npr # para generar numeros aleeatorios
import matplotlib.pyplot as plt #Para graficar
import scipy.stats as sps # distribuciones
from scipy.stats import poisson
````

````python
# Probabilidad de que hayan 0 errores en una pagina
# x=0, lambda=1.5
P0 = poisson.pmf(0, 1.5)
print("la probabilidad de que hayan 0 errores en una pagina es de", P0)

# Probabilidad de que hayan 2 o menos errores en una pagina

P_2 = poisson.cdf(2, 1.5)
print("la probabilidad de que hayan 2 o menos errores en una pagina es de", P_2)
````

Realizamos la grafica:


````python
# Parámetro
lambda4 = 1.5
#Valores posibles de X
X4 = np.arange(0, 10)
#Probabilidad
prob4 = poisson.pmf(X4, lambda4)
#Grafica
plt.figure(figsize=(10,6))
plt.subplot(2, 2, 4)
plt.bar(X4, prob4, color='#E84393', edgecolor='black')
plt.title('Distribución de Poisson (λ=1.5)')
plt.xlabel('Errores por página')
plt.ylabel('Probabilidad')
plt.grid(True)
#Volvemos a colocar las probabilidades, pero ahora que se vean en la grafica
P_X_0_4 = poisson.pmf(0, lambda4)
P_X_leq_2_4 = poisson.cdf(2, lambda4)
print(f'Ejercicio 4: P(X=0)={P_X_0_4:.4f}, P(X<=2)={P_X_leq_2_4:.4f}')

plt.tight_layout()
plt.show()
````

**Interpretación:**

Sabemos que la variable aleatoria $X$ representa el número de errores en una página de un libro.

La probabilidad de que una página no tenga errores es:

$$
P(X=0)=0.2231
$$

Esto significa que aproximadamente el 22.31\% de las páginas no tienen errores.

Por otro lado, la probabilidad de que haya como máximo 2 errores es:

$$
P(X \leq 2)=0.8088
$$

lo que significa que en alrededor del 80.88\% de las páginas habrá 2 errores o menos.
