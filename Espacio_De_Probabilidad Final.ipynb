**Experimento Aleatorio**

Un experimento aleatorio es un experimento $$E$$ que reproducido en condiciones identicas puede conducir a varios resultados, en el cual no se puede predecir el resultado.

**Espacio Muestral**

El espacio de todos los posibles resultados de un experimento aleatorio se llama **espacio muestral** y se denota por

$$
\Omega.
$$

Un posible resultado del experimento aleatorio se denota por

$$
\omega,\qquad \omega \in \Omega.
$$
**Algunos ejemplos**

- Lanzamiento de una moneda

$$
\Omega=\{\text{águila},\ \text{sol}\}.
$$

- Lanzamiento de un dado

$$
\Omega=\{1,2,3,4,5,6\}.
$$

- Lanzamiento de un dardo sobre un círculo de 30 cm de diámetro

$$
\Omega=\left\{(x,y):\sqrt{x^2+y^2}\le 15\right\}.
$$

- Romeo espera a Julieta, donde el tiempo de espera puede ser desde 1 minuto hasta 1 hora

$$
\Omega=[0,1].
$$

- Tiempo de vida de una bacteria

$$
\Omega=[0,\infty).
$$

**Evento Aleatorio**

Un evento aleatorio (asociado al experimento $E$) es un subconjunto del espacio de estados $\Omega$, para el cual se puede determinar si ocurre o no cuando se realiza el experimento.

Un evento aleatorio $$A$$ satisface:

$$
A \subseteq \Omega.
$$

**Algunos ejemplos:**

- Obtener sol al lanzar una moneda

$$
A=\{\text{sol}\}.
$$

- Obtener un 3 al lanzar un dado

$$
A=\{3\}.
$$

- ulieta se retrasa menos de un cuarto de hora

Si el tiempo se mide en horas,

$$
A=\left[0,\frac{1}{4}\right].
$$

En general, un evento aleatorio está representado por un conjunto de resultados para los cuales dicho evento ocurre.

**Espacio de probabilidad**

Un espacio de probabilidad es una terna

$$
(\Omega,\mathcal{F},P),
$$

donde:

- $\Omega$ es un conjunto no vacío, llamado espacio de estados.
- $\mathcal{F}$ es una familia de subconjuntos de $\Omega$, llamados eventos.
- $P$ es una medida de probabilidad, la cual asigna un número no negativo a cada evento.


**La medida de probabilidad** es una función

$$
P:\mathcal{F}\longrightarrow[0,1],
$$


que satisface los siguientes axiomas:

1. Para todo $A\in\mathcal{F}$,

$$
0\leq P(A)\leq 1.
$$

2. La probabilidad del espacio de estados es igual a 1, es decir,

$$
P(\Omega)=1.
$$

3. Si $(A_n)_{n\geq1}$ es una sucesión finita o infinita numerable de elementos de $\mathcal{F}$, mutuamente excluyentes dos a dos (es decir,

$$
A_i\cap A_j=\varnothing,\qquad i\neq j,
$$

entonces se cumple la propiedad de aditividad numerable:

$$
P\left(\bigcup_{n=1}^{\infty}A_n\right)
=
\sum_{n=1}^{\infty}P(A_n).
$$

**Propiedades Elementales**

Sea $(\Omega,\mathcal{F},P)$ un espacio de probabilidad. Entonces:

1.

$$
P(\varnothing)=0.
$$

2. Si $A_1,\ldots,A_n\in\mathcal{F}$ son ajenos dos a dos, entonces

$$
P\left(\bigcup_{k=1}^{n}A_k\right)
=
\sum_{k=1}^{n}P(A_k).
$$

3.

$$
P(A^c)=1-P(A).
$$

4. Si $A\subseteq B$, entonces

$$
P(B-A)=P(B)-P(A).
$$

5. Si $A\subseteq B$, entonces

$$
P(A)\leq P(B).
$$

6.

$$
0\leq P(A)\leq 1.
$$

7.

$$
P(A\cup B)=P(A)+P(B)-P(A\cap B).
$$

8.

$$
P(A\cup B)\leq P(A)+P(B).
$$

**Probabilidad Condicional**


Si añadimos condiciones a nuestro experimento aleatorio, la probabilidad cambia.

Definimos  la probabilidad de que ocurra un evento $A$ **sabiendo** que ocurre un evento $B$, de la siguiente manera:

$$\mathbb{P}(A \mid B) = \frac{\mathbb{P}(A \cap B)}{\mathbb{P}(B)} = \frac{|A \cap B|}{|B|},$$
donde $\mathbb{P}(B)>0$.




```python
# Definimos la funcion de probabilidad condicional
def prob_cond(E,F):
  return len(E & F) / len(F)
```

**Ejemplo**

Tenemos una urna con las siguientes bolas:

- 5 bolas rojas.
- 3 bolas azules.
- 2 bolas verdes.

Entonces en total hay:

$$
5+3+2=10 \text{ bolas}
$$

Se extraen dos bolas sin reemplazo.

**a)** Calcular la probabilidad de que la primera bola sea roja.

Definimos el evento

$$
A=\{\text{La primera bola extraída es roja}\}
$$

Como existen 5 bolas rojas de un total de 10 bolas, utilizando el enfoque clásico de la probabilidad obtenemos

$$
P(A)=\frac{\text{Casos favorables}}{\text{Casos posibles}}
=\frac{5}{10}
=\frac12
$$

Por lo tanto,

$$
P(A)=0.5
$$


**b)** Si la primera bola fue roja, calcular la probabilidad de que la segunda también sea roja.

Definimos el evento

$$
B=\{\text{La segunda bola extraída es roja}\}
$$

Como sabemos que la primera bola fue roja, queremos calcular la probabilidad de que la segunda también sea roja bajo esa condición. Para ello utilizamos la definición de probabilidad condicional:

$$
P(B\mid A)=\frac{P(A\cap B)}{P(A)}
$$

Donde:

- $P(B\mid A)$ es la probabilidad de que la segunda bola sea roja dado que la primera fue roja.
- $P(A\cap B)$ representa la probabilidad de que ambas bolas sean rojas.
- $P(A)$ es la probabilidad de que la primera bola sea roja.

Calculamos cada una de estas probabilidades.

La probabilidad de que la primera bola sea roja es

$$
P(A)=\frac{5}{10}
$$

Por otra parte, para que ambas bolas sean rojas, primero debe salir una de las 5 bolas rojas y después una de las 4 bolas rojas restantes, ya que la extracción es sin reemplazo. Entonces

$$
P(A\cap B)=\frac{5}{10}\cdot\frac{4}{9}
=\frac{2}{9}
$$

Sustituyendo en la fórmula de probabilidad condicional,

$$
P(B\mid A)
=\frac{\frac{2}{9}}{\frac{5}{10}}
$$

Simplificando,

$$
P(B\mid A)
=\frac{2}{9}\cdot\frac{10}{5}
=\frac{20}{45}
=\frac49
$$

Es decir

$$
P(B\mid A)=0.4444
$$



```python
# Importamos la librería NumPy para generar números aleatorios
import numpy as np

# Fijamos una semilla para que los resultados sean reproducibles
np.random.seed(123)

# Definimos el número de simulaciones
n = 100000

# Construimos la urna:
# R = Roja, A = Azul, V = Verde
urna = ['R'] * 5 + ['A'] * 3 + ['V'] * 2

# Inicializamos un contador para registrar las veces que la primera bola es roja
primera_roja = 0

# Inicializamos un contador para registrar las veces que ambas bolas son rojas
ambas_rojas = 0

# Realizamos las simulaciones
for i in range(n):

    # Extraemos dos bolas sin reemplazo
    extraccion = np.random.choice(urna, size=2, replace=False)

    # Verificamos si la primera bola es roja
    if extraccion[0] == 'R':

        # Contamos el evento A: la primera bola es roja
        primera_roja += 1

        # Verificamos si la segunda bola también es roja
        if extraccion[1] == 'R':

            # Contamos el evento A∩B: ambas bolas son rojas
            ambas_rojas += 1

# Calculamos la probabilidad simulada de que la primera bola sea roja
prob_a = primera_roja / n

# Calculamos la probabilidad condicional simulada P(B|A)
prob_b_dado_a = ambas_rojas / primera_roja

# Mostramos la probabilidad teórica y simulada del inciso a)
print("Respuesta al inciso a)")
print("Probabilidad teórica P(A) =", 5/10)
print("Probabilidad simulada P(A) =", prob_a)



# Mostramos la probabilidad teórica y simulada del inciso b)
print("Respuesta al inciso b)")
print("Probabilidad teórica P(B|A) =", 4/9)
print("Probabilidad simulada P(B|A) =", prob_b_dado_a)
```

**Probabilidad Geometrica**




En el modelo geométrico de probabilidad, las probabilidades se obtienen comparando una longitud, un área o un volumen favorable con el total del espacio muestral.

En el caso de problemas de áreas, la probabilidad se calcula como:

$$
P(A)=\frac{\text{Área deseada}}{\text{Área total}}.
$$


**Problema del encuentro**

Dos personas planean encontrarse en un lugar especifico en un rango de una hora $[0,60]$ minutos (1 hora)
- Si A llega primero, entonces piensa esperar 10 minutos
- Si B llega primero, entonces piensa esperar 10 minutos
Se elige un numero al azar entre $[0,60]$ y supongamos que el par de numeros se elige sobre el cuadrado de modo que la probabilidad de un evento sea el area favorable entre el area total.

Sea

- $A$: las dos personas se encuentran a una distancia de 10 minutos.

Nos gustaria encontrar la probabilidad de $A$.


```python
import numpy as np
import matplotlib.pyplot as plt #para graficos

#vamos a generar 1000 simulaciones
muestra= 1000
#generamos los tiempos aleatorios entre [0,60]
x= np.random.uniform(0,60,muestra)
y= np.random.uniform(0,60,muestra)

#Añadimos una condicion de encuentro
encuentro= np.where(abs(x-y) <= 10, 'mediumseagreen', 'cornflowerblue') #verde si se encuentran, azul si no

#ahora calculamos la probabilidad de que se encuentren
k=0
for elemento in encuentro:
    if elemento == 'mediumseagreen': #l0s favorables
       k += 1
print("La probabilidad del evento A es: {}". format(k/muestra))

#graficamos
plt.figure(figsize=(5,5)) #tamaño de la figura
plt.scatter(x,y, c=encuentro, s=1)

#Ponemos el título de la gráfica
plt.title("Simulación del problema del encuentro")

# ponemos el nombre de los ejes
plt.xlabel("Tiempo de llegada de la persona A (minutos)")
plt.ylabel("Tiempo de llegada de la persona B (minutos)")

#Ponemos el limite de los ejes
plt.xlim(0,60)
plt.ylim(0,60)

#Establecemos la cuadricula
plt.grid(alpha=0.3)
#Mostramos la grafica
plt.show()
```
