**MUESTRA ALEATORIA**

Sean

$$
X_1,\ldots,X_N
$$

variables aleatorias independientes e idénticamente distribuidas. Al conjunto

$$
\{X_i\}_{i=1}^{N}
$$

donde

$$
i\in\{1,\ldots,N\}.
$$

Siempre que hagamos referencia a una muestra aleatoria (m.a.), pensaremos en un conjunto de variables que cumplen las condiciones iniciales.

**ESTIMADOR**

Se define un estimador $W$ como una transformación de la muestra aleatoria que no contiene el parámetro en cuestión y que además busca aproximar el valor del parámetro.

**METODO DE MOMENTOS**

El método de los momentos es uno de los procedimientos más antiguos para obtener estimadores puntuales de parámetros poblacionales. Fue propuesto por Karl Pearson a finales del siglo XIX y se caracteriza por ser un método sencillo de aplicar y por proporcionar estimadores en una gran variedad de problemas.

La idea principal consiste en igualar los momentos poblacionales, que dependen de los parámetros desconocidos, con los momentos muestrales, calculados a partir de una muestra aleatoria. A partir de estas igualdades se obtienen ecuaciones cuya solución proporciona los estimadores de los parámetros.

En muchas situaciones este método produce buenos estimadores y es un excelente punto de partida cuando otros métodos de estimación son difíciles de aplicar. Sin embargo, los estimadores obtenidos mediante el método de los momentos no siempre poseen las mejores propiedades estadísticas, por lo que en algunos casos pueden mejorarse mediante otros métodos, como el de máxima verosimilitud.

Sea

$$
X_1,\ldots,X_n
$$

una muestra de una población con función de densidad o función de probabilidad

$$
f(x\mid \theta_1,\ldots,\theta_k).
$$

Los estimadores por el **método de los momentos** se obtienen igualando los primeros $k$ momentos muestrales con los correspondientes $k$ momentos poblacionales y resolviendo el sistema de ecuaciones resultante.

Más precisamente, se definen

$$
m_1=\frac{1}{n}\sum_{i=1}^{n}X_i,
\qquad
\mu_1'=E(X),
$$

$$
m_2=\frac{1}{n}\sum_{i=1}^{n}X_i^2,
\qquad
\mu_2'=E(X^2),
$$

$$
\vdots
$$

$$
m_k=\frac{1}{n}\sum_{i=1}^{n}X_i^k,
\qquad
\mu_k'=E(X^k).
$$

El momento poblacional $\mu_j'$ será, en general, una función de los parámetros

$$
\theta_1,\ldots,\theta_k,
$$

es decir,

$$
\mu_j'=\mu_j'(\theta_1,\ldots,\theta_k).
$$

El estimador por el método de los momentos

$$
(\hat{\theta}_1,\ldots,\hat{\theta}_k)
$$

se obtiene resolviendo el siguiente sistema de ecuaciones:

$$
m_1=\mu_1'(\theta_1,\ldots,\theta_k),
$$

$$
m_2=\mu_2'(\theta_1,\ldots,\theta_k),
$$

$$
\vdots
$$

$$
m_k=\mu_k'(\theta_1,\ldots,\theta_k).
$$

**Estimación por máxima verosimilitud (EMV)**

El **método de máxima verosimilitud** es uno de los procedimientos más utilizados para obtener estimadores puntuales. La idea consiste en elegir el valor del parámetro que hace más probable la muestra observada.

Sea

$$
X_1,\ldots,X_n
$$

una muestra aleatoria proveniente de una población con función de densidad o función de probabilidad

$$
f(x\mid\theta).
$$

La **función de verosimilitud** se define como

$$
L(\theta\mid\mathbf{x})
=\prod_{i=1}^{n}f(x_i\mid\theta),
$$

cuando las observaciones son independientes.

La **estimación por máxima verosimilitud** se basa en encontrar el valor del parámetro $\theta$ para el cual la función de verosimilitud alcanza un máximo.

Para ello, se deriva la función de verosimilitud respecto al número de parámetros desconocidos.

$$
\frac{\partial}{\partial\theta_1}L(\theta\mid\mathbf{x}),
$$

$$
\frac{\partial}{\partial\theta_2}L(\theta\mid\mathbf{x})
$$

$$
\vdots
$$

$$
\frac{\partial}{\partial\theta_k}L(\theta\mid\mathbf{x})
$$

Posteriormente, procedemos a igualar a 0 las derivadas para ver si existe un punto critico y posterior a eso verificar la existencia del maximo. Es decir:
$$
\frac{\partial}{\partial\theta_1}L(\theta\mid\mathbf{x})=0,
$$

$$
\frac{\partial}{\partial\theta_2}L(\theta\mid\mathbf{x})=0,
$$

$$
\vdots
$$

$$
\frac{\partial}{\partial\theta_k}L(\theta\mid\mathbf{x})=0.
$$

**Ejemplo.**

Lanzamod una moneda 100 veces, sabiendo que se obtienen 55 soles. Encontrar el EMV para la probabilidad $p$ de obtener sol en un solo lanzamiento.

Sea
$X$: el número de soles que se obtienen al lanzar una moneda
se tiene que $X\sim \text{Bin}(100,p)$, entonces

$$ f(x;p) = \binom{100}{55}p^{55}(1-p)^{100-55}$$

```python
from scipy.stats import binom
import numpy as np

# Vamos a utilizar la libreria Sympy para matemática símbolica
import sympy

# importa la variable símbolica x
from sympy.abc import x
```

https://www.sympy.org/en/features.html

```python
# Vamos a definir a p como una variable símbolica positiva: 0<p<1
p = sympy.symbols('p', positive= True)

# Definimos la función de verosimilitus
# se puede omitir la constante
f = p**55 * (1-p)**45

# Ahora derivamos f con respecto a p
df = sympy.diff(f,p)

# Hay que igualar la derivada a cero y despejar a p
# df=0 para encontrar p
phat = sympy.solve(df,p)[0] # [1] es la segunda solucion y [0] es la primera

print("El estimador de máxima verosimilitud es", phat)
```

La cual coincide con la proporción de soles.

**Interpretacion:**

El estimador de máxima verosimilitud obtenido es

$$
\hat{p}=\frac{11}{20}=0.55
$$

Esto significa que con base en los datos observados, la mejor estimación para la probabilidad de obtener sol en un solo lanzamiento de la moneda es de 0.55. En otras palabras, se estima que la moneda tiene aproximadamente un 55% de probabilidad de caer del lado del sol en cada lanzamiento.

Este resultado coincide con la proporción de soles observada en la muestra, ya que de los 100 lanzamientos realizados se obtuvieron 55 soles.

\hat{\theta}=\frac{1}{\bar{X}}\approx 0.00275 \text{ por minuto}

**Insesgadez**

Un estimador se dice **insesgado** cuando su sesgo es igual a cero. Equivalentemente,

$$
E(\hat{\theta})=\theta.
$$

**Interpretación.**

Un estimador es insesgado si, en promedio, estima correctamente el valor del parámetro poblacional. Es decir, si se tomaran muchas muestras y se calculara el estimador en cada una de ellas, el promedio de esos estimadores sería igual al verdadero valor del parámetro.

**Eficiencia**

Diremos que dados dos estimadores $\hat{\theta}_1$ y $\hat{\theta}_2$, $\hat{\theta}_1$ es **más eficiente** respecto a $\hat{\theta}_2$ si
$$
\operatorname{Var}(\hat{\theta}_1)
<
\operatorname{Var}(\hat{\theta}_2).
$$

**Interpretación.**

Un estimador es eficiente si, entre todos los estimadores insesgados de un mismo parámetro, presenta la menor varianza. Esto significa que sus estimaciones tienden a estar menos dispersas y, por lo tanto, son más precisas.

**Consistencia**

Un estimador se dice **consistente** si se cumplen las siguientes condiciones:

$$
\lim_{n\to\infty}E(\hat{\theta}_n)=\theta,
$$

y

$$
\lim_{n\to\infty}\operatorname{Var}(\hat{\theta}_n)=0.
$$

**Interpretación.**

Un estimador es consistente si, conforme aumenta el tamaño de la muestra, el valor del estimador se aproxima cada vez más al verdadero valor del parámetro. En otras palabras, con muestras grandes el estimador tiende a ser más preciso.

**INTERVALOS DE CONFIANZA**

Un intervalo de confianza es un rango de valores que con cierta probabilidad (nivel de confianza) contiene al verdadero parámetro poblacional que queremos estimar.

Sea $X_1,\dots, X_n$ una muestra aleatoria con $X\sim\text{N}(\mu,\sigma^2)$, entonces
$$(num_1 < \mu < num2)$$

En los intervalos de confianza, tenemos el error estándar (SE) el cual mide cuanto varía un estimador.
 - si conocemos $\sigma$, entonces: $SE = \sigma/\sqrt{n}$
 - Si NO conocemos $\sigma$, utilizamos $s$ y entonces: $SE=\frac{s}{\sqrt{n}}$

NOTA: a mayor tamaño de la muestra se tiene que el SE será menor y por lo tanto el intervalo de confianza será más preciso.

**Def** Si $X_1,\dots, X_n$ es una muestra aleatoria con densidad $f(x;\theta)$ y sean $T_1=t_1(X_1,\dots, X_n)$ y $T_2=t_2(X_1,\dots, X_n)$ dos estadísticos talque $T_1 \le T_2$ entonces
$$\mathbb{P}(T_1 \le \theta \le T_2) = \gamma$$
Al intervalo $(T_1,T_2)$ se le llama intervalo del $\gamma\times 100\%$ de confianza.


**Intervalo de confianza para la media de una población normal**

Sea $X_1,\dots, X_n$ una muestra aleatoria con $X\sim\text{N}(\mu,\sigma^2)$, entonces
 - $\bar{X_i} \sim \mathcal{N}(\mu, \sigma^2/n)$
 - $Z = \frac{\bar{X_i}-\mu}{\sigma/n} \sim \mathcal{N}(0,1)$

**$\sigma$ conocida:**

Se tiene que el intervalo del $(1-\alpha)\times 100\%$ de confianza para $\mu$ es:
$$ \bar{X}\pm z_{\alpha/2}\frac{\sigma}{\sqrt{n}}$$
donde
 - $z_{\alpha/2}\frac{\sigma}{\sqrt{n}}$ es el margen de error
 - $z_{\alpha/2}$ es el valor crítico

**Ejemplo** Para tratar de estimar la media de consumo por cliente en un restaurante, se reunieron datos de una muestra de tamaño $49$ clientes durante tres semanas.

**a)** Supongamos que la desviación estándar poblacional es de $\$2.5$. ¿Cuál es el error estandar de la media?

**b)** Con un nivel de confianza del $95\%$, ¿cuál es el margen de error?

**c)** Si la media muestral es de $\$22.60$, cual es el intervalo de confianza del $95\%$ para la media poblacional.

```python
import numpy as np
import scipy.stats as st

# Datos:
n = 49 # número de clientes
sigma = 2.5
media_muestral = 22.60
confianza = 0.95

# a) SE
error_est = sigma / np.sqrt(n)
print('El error estándar es:', error_est)

# b) 1-confianza = 0.95 -> 0.025 en cada cola -> (1-confianza)/2
z = st.norm.ppf(1-(1-confianza)/2)
margen_error = z*error_est
print('El margen de error es:', margen_error)

# c) Intervalo de confianza
limite_inf = media_muestral - margen_error
limite_sup = media_muestral + margen_error
print(f'El intervalo del {confianza*100}% es: ({limite_inf:.2f},{limite_sup:.2f})')
```

Por lo tanto, se tiene que el verdadero consumo poblacional en el restaurante se encuentra entre $\$21.90$ y $\$23.30$

El intervalo de confianza del 95% para la media del consumo por cliente es

$$
(21.90,\;23.30).
$$

Esto significa que, con un nivel de confianza del 95%, se estima que la media poblacional del consumo por cliente en el restaurante se encuentra entre 21.90 y 23.30. Este intervalo fue construido a partir de la información obtenida de una muestra de 49 clientes.
En otras palabras, si se tomaran muchas muestras del mismo tamaño y se construyera un intervalo de confianza del 95% para cada una de ellas, aproximadamente el 95% de esos intervalos contendrían el verdadero valor de la media poblacional.

**$\sigma$ desconocida:**

Se tiene que el intervalo del $(1-\alpha)\times 100\%$ de confianza para $\mu$ es:
$$ \bar{X}\pm t_{\alpha/2}\frac{\sigma}{\sqrt{n}}$$
donde
 - $t_{\alpha/2}\frac{\sigma}{\sqrt{n}}$ es el margen de error
 - $t_{\alpha/2}$ es el valor crítico

**Ejemplo**  

Supongamos que tenemos los siguientes datos de una muestra aleatoria de 10 observaciones:
$$45,55,67,68,45,79,98,87,84,82$$

Calcular un intervalo de confianza para la media.

```python
# Datos de la muestra (n ≤ 30)
datos = [45,55,67,68,45,79,98,87,84,82]
n = len(datos)
confianza = 0.99
gl = n - 1

# Calculamos la media muestral
media_muestral = np.mean(datos)

# Calculamos el error estándar de la media
error_est = st.sem(datos)

# Utilizamos la función:
# scipy.stats.t.interval(confidence, df, loc, scale)
# confidence = nivel de confianza
# df = grados de libertad
# loc = media muestral
# scale = error estándar

intervalo = st.t.interval(confianza, gl, media_muestral, error_est)

print(f'El intervalo de confianza del {confianza*100:.0f}% es: ({intervalo[0]:.2f}, {intervalo[1]:.2f})')
```

El intervalo de confianza del 99% para la media poblacional es

$$
(52.32,\;89.68).
$$

Esto significa que, con un nivel de confianza del 99%, se estima que la media poblacional de la variable de interés se encuentra entre 52.32 y 89.68 Debido a que el tamaño de la muestra es pequeño n=10 y la desviación estándar poblacional es desconocida, el intervalo se construyó utilizando la distribución t de Student.

En otras palabras, si se obtuvieran muchas muestras aleatorias del mismo tamaño y se construyera un intervalo de confianza del 99% para cada una de ellas, aproximadamente el 99% de esos intervalos contendrían el verdadero valor de la media poblacional.

**Intervalo de confianza para la varianza de una población normal**

**$\mu$ desconocida:**

Se tiene que el intervalo del $(1-\alpha)\times 100\%$ de confianza para $\sigma^2$ es:
$$ \left( \frac{(n-1)s^2}{\chi^{2}_{1-\alpha/2}}, \frac{(n-1)s^2}{\chi^{2}_{\alpha/2}} \right)$$

**Ejemplo**

Los artículos de cerámica utilizados sobre velas electricas se rompen con diferentes presiones. Supongamos que los datos provienen de una distribución normal. La resistencia a la ruptura fue medida en una muestra de 100 artículos y el promedio fue de $1750$ con una desviación estándar de $315.8$.

**a)** Estimar con un nivel de confianza del $90\%$ a la media poblacional de la presión de la ruptura.

**b)** Estimar con un nivel de confianza del $90\%$ a la varianza poblacional.

```python
# Datos
n = 100
confianza = 0.90
media_muestral = 1750
desv_est = 315.8
error_est = desv_est / np.sqrt(n)

# a) como n>=30, utilizo la normal
intervalo = st.norm.interval(confianza, media_muestral, error_est)
print(f'El intervalo de confianza del {confianza*100}%: ({intervalo[0]:.2f},{intervalo[1]:.2f})')

# b) IC para la varianza
gl = n-1
alpha = 1-confianza
s2 = desv_est**2 # varianza muestral

# Forma 1. Calculando los cuatiles  de la chi cuadrada
chi_inf = st.chi2.ppf(alpha/2,gl)
chi_sup = st.chi2.ppf(1-alpha/2,gl)
int_var = ((gl*s2) / chi_sup, (gl*s2) / chi_inf)
print(f'El intervalo de confianza del {confianza*100}%: ({int_var[0]:.2f},{int_var[1]:.2f})')

# Forma 2.
chi2_inf, chi2_sup = st.chi2.interval(confianza, gl)
int2_var = ((gl*s2) / chi2_sup, (gl*s2) / chi2_inf)
print(f'El intervalo de confianza del {confianza*100}%: ({int2_var[0]:.2f},{int2_var[1]:.2f})')
```

El intervalo de confianza del 90% para la media poblacional de la presión de ruptura es

$$
(1698.05,\;1801.95)
$$

Esto significa que, con un nivel de confianza del 90%, se estima que la media poblacional de la presión necesaria para romper los artículos de cerámica se encuentra entre 1698.05 y 1801.95 unidades de presión.

En otras palabras, si se tomaran muchas muestras de 100 artículos y se construyera un intervalo de confianza del 90% para cada una de ellas, aproximadamente el 90% de esos intervalos contendrían el verdadero valor de la media poblacional.

Esto significa que, con un nivel de confianza del 90%, se estima que la varianza poblacional de la presión de ruptura se encuentra entre 80123.49 y 128146.72unidades al cuadrado. Este intervalo refleja la variabilidad esperada de la presión de ruptura de los artículos de cerámica en la población.

En otras palabras, si se obtuvieran muchas muestras del mismo tamaño y se construyera un intervalo de confianza del 90% para la varianza en cada una de ellas, aproximadamente el 90% de esos intervalos contendrían el verdadero valor de la varianza poblacional.

**Intervalo de confianza para una proporción**

Cuando se desea estimar la proporción poblacional $p$ a partir de una muestra aleatoria de tamaño $n$, se utiliza la proporción muestral

$$
\hat{p}=\frac{x}{n},
$$

donde $x$ representa el número de elementos de la muestra que poseen la característica de interés.

Si el tamaño de la muestra es suficientemente grande, un intervalo de confianza de nivel $1-\alpha$ para la proporción poblacional está dado por

$$
\hat{p}\pm z_{\alpha/2}\sqrt{\frac{\hat{p}(1-\hat{p})}{n}}.
$$

donde:

- $\hat{p}$ es la proporción muestral.
- $z_{\alpha/2}$ es el valor crítico de la distribución normal estándar asociado al nivel de confianza.
- $n$ es el tamaño de la muestra.

Este intervalo proporciona un rango de valores dentro del cual se espera que se encuentre la proporción poblacional con el nivel de confianza especificado.

**Ejemplo**

En una encuesta de una Universidad, 225 estudiantes son seleccionados al azar de los que toman cálculo y se les pregunta si cálculo es su materia favorita. 100 estudiantes responden que el cálculo es su materia favorita. Proporcione un intervalo de confianza del 95 %  para la proporción de todos los estudiantes que toman cálculo y que la consideran su materia favorita.  

```python
# Datos
n = 225
fav_cal = 100
confianza = 0.95

# Calculamos la Proporción
prop_muestral = fav_cal / n

# Calculamos el error estándar
error_est = np.sqrt((prop_muestral * (1 - prop_muestral)) / n)

# Calculamos el valor crítico
z = st.norm.ppf(1 - (1 - confianza) / 2)

# Calculamos el margen de error
margen_error = z * error_est

# Ya teniendo todo, construimos el intervalo de confianza
limite_inf = prop_muestral - margen_error
limite_sup = prop_muestral + margen_error

print(f'El intervalo de confianza del {confianza*100}% es: ({limite_inf:.4f}, {limite_sup:.4f})')
```

En este ejercicio se construyó un intervalo de confianza para estimar una proporción poblacional a partir de una muestra. Debido a que el tamaño de la muestra es grande, se utilizó la aproximación mediante la distribución normal Z.

El intervalo de confianza del 95% indica que la proporción de estudiantes que consideran Cálculo como su materia favorita se encuentra entre 37.95 %y 50.93.

En otras palabras, si se seleccionaran muchas muestras aleatorias de 225 estudiantes y se construyera un intervalo de confianza del 95% para cada una de ellas, aproximadamente el 95% de esos intervalos contendrían la verdadera proporción de estudiantes que consideran Cálculo como su materia favorita.
