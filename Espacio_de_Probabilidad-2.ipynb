**Probabilidad Laplaciana**
Pra un experimento aleatorio, se tiene que $\Omega$ es el espacio muestral (conjunto de popsibles resultados).
Sea $A$ un evento, entonces definimos la probabilidad de $A$ como sigue:
$$ \mathbb{P} (A)= \frac{|A|}{|\Omega|}. $$
 Ejemplo sea el experimento de lanzar una moneda 3 veces consecutivas, de manera independiente, es decir , el $i$ esimo lanzamiento no afecta al $i+1$ lanzamiento. Es espacio muestral es:
 $$ \Omega_{\text{moneda}}= \{a,s\}$$

 $$ \Omega = \Omega_{\text{moneda}}\times \Omega_{\text{moneda}}\times \Omega_{\text{moneda}} =\{(s,s,s),(s,a,s),(s,s,a),(s,a,a),(a,s,a),(a,a,s),(a,a,a)\} $$
 y se tiene que $|\Omega|= 2 ala 3= 8$
  sean los eventos:
  - $A$: todos los resultados son aguila.
  - $S$: todos los resultados son son
  - $B$: por lo menos un resultado es sol (1 o 2 o soles)
  vamos a calcular las probabilidades de $ A$ , $S$ y $B$. Para esto sea
  - $ A_1$: cae agulia en el $i$- esimo lanzamiento.
  Para el evento $A$, se tiene que $A=A_1\ cap A_2 cap A_3

**PROBABILIDAD CONDICIONAL**


Definimos la probabilidad condicional de que ocurra un evento $A$ sabiendo que ocurre un evento $B$ de la siguiente manera:
$$ \mathbb{P}(A \mid B) =$$frac{\mathbb{P}(A\cap B)}{\mathbb{P}(B)} = \frac{|A\cap B| checar video xd





```python
# Definimos la funcion de probabilidad condicional
def prob_cond(E,F):
  return len(E & F) / len(F)
```
