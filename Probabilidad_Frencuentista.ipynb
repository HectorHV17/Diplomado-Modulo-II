Sea un espacio muestral (conjunto finito) podemos definir la probabilidad aosciada a un evento $A$ donde tengo un experiemento aleatorio que se realiza $n$ veces

$$ f_n(A) = \frac{n(A)}{n} $$
donde $n(A)$ es el número de veces en que ocurre $A$.


Sea el experimento aleatorio de lanzar una moneda (no junta).

Vamos a simular la probabilidad frecuentista, es decir, vamos a realizar un experimento con 1000 volados (individual) y puedo repetir 500 veces (500 personas)

```python
#Creamos una lista para almacenar las probabilidades
p=[]
#Reañiza 500 experimentos
#cada expe tiene 1000 voladps
#lo almacena sol y aguila
for i in range(500):
    N=1000
    resultados=[]

    for j in range(N):
        resultado= volado(0.63)
        resultados.append(resultado)

    n_soles=resultados.count("sol")
    p_soles= n_soles/ N

    #Añadimos estas probabilidades a la lista P
    p.append(p_soles)
    #probabilidad frec.promedio
    p_soles_prom= sum(p)/500

p_soles_prom
```

EJEMPLO: Problema del encuentro
Dos personas planean encontrarse en un lugar especifico en un rango de una hora $[0,60]$ minutos (1 hora)
- Si A llega primero, entonces piensa esperar 10 minutos
- Si B llega primero, entonces piensa esperar 10 minutos
Se elige un numero al azar entre $[0,60]$ y supongamos que el par de numeros se elige sobre el cuadrado de modo que la probabilidad de un evento sea el area favorable entre el area total.

Sea

- $A$: las dos personas se encuentran a una distancia de 10 minutos.

Nos gustaria encontrar la probabilidad de $A$.


```python
import numpy as np
import matplotlib.pyplot as plt #para graficos

#vamos a generar 1000 simulaciones
muestra= 1000
#generamos los tiempos aleatorios entre [0,60]
x= np.random.uniform(0,60,muestra)
y= np.random.uniform(0,60,muestra)

#Añadimos una condicion de encuentro
encuentro= np.where(abs(x-y) <= 10, 'r', 'b') #rojo si se encuentrajn, azul si no

#ahora calculamos la probabilidad de que se encuentren
k=0
for elemento in encuentro:
    if elemento == 'r': #l0s favorables
       k += 1
print("La probabilidad del evento A es: {}". format(k/muestra))

#graficamos
plt.figure(figsize=(5,5)) #tamaño de la figura
plt.scatter(x,y, c=encuentro, s=1)
plt.show
```
